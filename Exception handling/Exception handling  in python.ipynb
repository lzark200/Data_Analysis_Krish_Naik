```python
try:
    num=int(input())
    res = 10/num
    print(res)
except ValueError as ve : 
    print(ve , "This is not a valid number")
except Exception as e : 
    print(e)
```

```python
## try-except, else-block : 
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError as ve:
    print(ve)
except ZeroDivisionError as zde:
    print(zde)
else:
    print(f"The result is {result}")
```

```python
## nested try-catch : 
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError as ve:
    print(ve)
except ZeroDivisionError as zde:
    try:
        print(zde+5)
    except Exception as e:
        print(e)
else:
    print(f"The result is {result}")
```

```python
## try, except, else and finally
## try-except, else-block : 
try:
    num = int(input("enter a number"))
    result = 10/num
except ValueError as ve:
    print(ve)
except ZeroDivisionError as zde:
    print(zde)
else:
    print(f"The result is {result}")
finally:
    print("Exceution complete")
```

finally is the block of the code that exceutes in all the cases either error is comming or the error is not comming

### Practicle cases in which I can apply exception handling : 

```python
try:
    file = open("file.txt" , 'r')
    content = file.read()
    print(content)
except Exception as e :
    print(e , "The file does not exist")
else:
    print("file successfully opened and has been read")
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("file closed successfully")
    else:
        print('file is either closed or does not exist')
    
```
